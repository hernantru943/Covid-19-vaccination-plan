# Covid-19 Vaccination plan (MVP)

The main objetive of this project is to deliver a preliminary analysis for a non profit organization called Foundation for Morristown Medical Center, who will improve tremendously the effectiveness of its resources by organizing a logistic of their "Vaccination points" in NY by identifying the bussiest stations by day and hour to get as many vaccinated population as possible in low income communities.

During the initial cleaning and exploration of the MTA data, we were able to understand the general behavior of their MTA system. We found out, as a regular big city, one of the bussiest stations shows a very high traffic from Monday through Friday so in order to get as many vaccinated people as possible, our client must be intall its vaccinations points during week days. 

Also, our initial findings showed us the 10 bussiest stations of the MTA system during the 11 weeks of our analysis which will be key for the logistic purposes towards our final goal.

We will definitely go further on our analysis, but this clouple graphics will give to our client a good sense of our work and how Data Science can help us on several ways such as loggistic improvement as well as the use of the resources.
